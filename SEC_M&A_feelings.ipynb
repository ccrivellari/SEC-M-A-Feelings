```
#retrieving SEC filings DEFM14A forms via SEC API of the arranged deal considering the selected basket of securities.

#package used to execute HTTP POST request to the API
import json
import urllib.request
import numpy as np
import pandas as pd

# API Key # replace YOUR_API_KEY with the API key you got from sec-api.io after sign up
TOKEN = '587153fa876792e8026ed84892ae68b2d0529b4b520444953ee896b137fa6278'
# API endpoint
API = "https://api.sec-api.io?token=" + f"{TOKEN}"
 
# define the filter parameters you want to send to the API
# CIK: 1279363(WCG US), 1578845 (AGN US), 1447526 (CZR US), 1489136 (SEMG US), 1095996 (WLH US), 1668428 (FG US)


for k, v in df_limited['Target CIK'].iteritems():

 payload = {
    "query": { "query_string": { "query": "cik: 1122388 AND filedAt:{2019-01-01 TO 2020-09-30} AND formType:\"DEFM14A\"" } },
    #"query": { "query_string": { "query": "cik: f\"{v}\" AND filedAt:{2019-01-01 TO 2020-09-30} AND formType:\"DEFM14A\"" } },
    "from": "0",
    "size": "1",
    "sort": [{ "filedAt": { "order": "desc" } }]
 }


# format your payload to JSON bytes
jsondata = json.dumps(payload)
jsondataasbytes = jsondata.encode('utf-8')   # needs to be bytes
 
# instantiate the request 
req = urllib.request.Request(API)
 
# set the correct HTTP header: Content-Type = application/json
req.add_header('Content-Type', 'application/json; charset=utf-8')
# set the correct length of your request
req.add_header('Content-Length', len(jsondataasbytes))
 
# send the request to the API
response = urllib.request.urlopen(req, jsondataasbytes)
 
# read the response 
res_body = response.read()
# transform the response into JSON
secdocs = json.loads(res_body.decode("utf-8"))
 
# print JSON 
print(secdocs)

dfsec = pd.pandas.json_normalize(secdocs, ['filings'])
dfsec.to_csv('DEFM14A' f"{dfsec['ticker']}"+'.csv')
```
